```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
base_data = pd.read_excel("/content/base alunos.xlsx")
```

```python
priv_data = base_data[base_data['TipoEsc'] == 'priv']
public_data = base_data[base_data['TipoEsc'] == 'pub']

def calcular_estatisticas(data, tipo_escola):
    quantidade_alunos = data.shape[0]
    media_notas = data['Nota_Teste'].mean()
    desvio_padrao = data['Nota_Teste'].std()
    mediana = data['Nota_Teste'].median()
    nota_minima = data['Nota_Teste'].min()
    nota_maxima = data['Nota_Teste'].max()

    print(f"Estatísticas para escolas {tipo_escola}:")
    print(f"Quantidade de alunos: {quantidade_alunos}")
    print(f"Média das notas: {media_notas:.2f}")
    print(f"Desvio padrão: {desvio_padrao:.2f}")
    print(f"Mediana: {mediana}")
    print(f"Nota mínima: {nota_minima}")
    print(f"Nota máxima: {nota_maxima}")
    print()

calcular_estatisticas(priv_data, "Privadas")
calcular_estatisticas(public_data, "Públicas")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='TipoEsc', y='Nota_Teste', data=base_data, color='aqua')
plt.title('Boxplot das Notas de Teste por Tipo de Escola')
plt.xlabel('Tipo de Escola')
plt.ylabel('Notas de Teste')
plt.grid(True)
plt.show()
```

**Conclusão:**
As estatísticas mostram que os alunos das escolas privadas, em geral, têm um desempenho acadêmico superior em comparação com os alunos das escolas públicas. A média mais alta e o menor desvio padrão das escolas privadas sugerem um desempenho mais consistente e geralmente melhor. Por outro lado, a maior variação nas notas e a menor média nas escolas públicas indicam desafios maiores em termos de desempenho acadêmico. Esses resultados podem ser influenciados por diversos fatores, incluindo recursos disponíveis, qualidade de ensino, e suporte educacional, entre outros.
